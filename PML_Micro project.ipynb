```python
month_map = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'June': 6,
             'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
osi['Month'] = osi['Month'].map(month_map)
```

```python
osi['VisitorType'] = osi['VisitorType'].map({'Returning_Visitor': 0, 'New_Visitor': 1, 'Other': 2})
```

##### 

```python
X=osi.drop(["Revenue"],axis=1)
y=osi["Revenue"].astype(int)
```

#    

### LinearSVC Model

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y,test_size=.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler as ss
std=ss()
ss_train=std.fit_transform(X_train)
ss_test=std.fit_transform(X_test)
```

```python
from sklearn.svm import LinearSVC as ls
from sklearn.metrics import *
lsvc=ls()
lsvc.fit(ss_train,y_train)
```

```python
y_pred=lsvc.predict(ss_test)
```

```python
SVC=accuracy_score(y_test,y_pred)
SVC
```

```python
print(classification_report(y_test,y_pred))
```

#   

### SGDClassifier

```python
from sklearn.linear_model import SGDClassifier as sgd
```

```python
sgd_model = sgd()
sgd_model.fit(ss_train, y_train)
sgd_y_pred = sgd_model.predict(ss_test)

SGD=accuracy_score(y_test,sgd_y_pred)
print('Accuracy Score for SGDClassifier model:',SGD)
print('\n')
print('Classification report for SGDClassifier model:')
print(classification_report(y_test, sgd_y_pred))
```
